```python
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from ml_lib import PCA as PCAfromTemu
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = load_breast_cancer()
X, y = dataset.data, dataset.target

tpca = PCAfromTemu(n_components=2)
pca = PCA(n_components=2)
preprocessor = StandardScaler()

X = preprocessor.fit_transform(X)

X_tpca = tpca.fit_transform(X)
X_pca = tpca.fit_transform(X)

```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

sns.scatterplot(x=X_tpca[:, 0], y=X_tpca[:, 1], hue=y, ax=ax1, palette='Set1')
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, ax=ax2, palette='Set1')

ax1.set_title("PCAfromTemu")
ax2.set_title("SklearnPCA")

plt.tight_layout()
```
