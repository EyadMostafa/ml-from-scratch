```python
from mpl_toolkits.mplot3d import Axes3D

X = 2 * np.random.rand(100, 2)
y = 2 * X[:, 0] + X[:, 1] + 3 + 0.5 * np.random.randn(100)

lr.fit(X, y)

x0 = np.linspace(X[:, 0].min(), X[:, 0].max(), 30)
x1 = np.linspace(X[:, 1].min(), X[:, 1].max(), 30)
x0_grid, x1_grid = np.meshgrid(x0, x1)
X_grid = np.c_[x0_grid.ravel(), x1_grid.ravel()]
y_pred = lr.predict(X_grid).reshape(x0_grid.shape)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[:, 0], X[:, 1], y, color="blue", label="Data")
ax.plot_surface(x0_grid, x1_grid, y_pred, color="red", alpha=0.5)
ax.set_xlabel("X0")
ax.set_ylabel("X1")
ax.set_zlabel("y")
plt.show()
```
